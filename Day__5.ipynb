```python
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
response
```

```python
import json
import pandas as pd
title_ = []
data_ = {'title':[],'createdAt':[],'commentCount':[],'likeCount':[]}
data = json.loads(response)
like_ = 0
comment_=0
# 計算熱門以及非熱門文章的平均留言人數和按讚人數
for d in data:
    title = d['title']
    title_.append(title)
   
    data['title'].append(str(d['title']))
    data['createdAt'].append(d['createdAt'])
    data['commentCount'].append(d['commentCount'])
    data['likeCount'].append(d['likeCount'])
   
    print(d['title'],'||',d['createdAt'],'||',d['commentCount'],'||',d['likeCount'])
    
```
