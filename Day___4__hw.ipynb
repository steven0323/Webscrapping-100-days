http 定義客戶端以及伺服器端如何傳輸資料
http request has two different approaches - get%post
4. get 會將要傳送的資料以query string 接在網址後面(key,value),而post則寫在封包裡面(form)

```python
import requests

r = requests.get('https://api.github.com/events')
r.text
```

```python
import json
json.loads(r.text)
```

```python
payload = {'key1':'value1','key2':'value2'}
r = requests.get("http://httpbin.org/get",params=payload)
json.loads(r.text)
```

```python
r = requests.post('http://httpbin.org/post',data = {''})
```
